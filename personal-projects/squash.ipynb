```python
#note, also followed multiple youtube tutorials however borrowed from different ones since I was not familiar with classes at the time I was working on this 
#Practice with turtles, actiononclick, and creating visual output that the user can interact with 
import turtle 

window=turtle.Screen()
window.title("Squash by Claire O'Neill")
window.bgcolor('#856ff8')
window.setup(width=800,height=600)
window.tracer(0) #prevents from running multiple times, makes program faster

# Score 
score_A=0
score_B=0
    
# Paddle A 
paddle_A=turtle.Turtle() #turtle object 
paddle_A.speed(0) #sets speed to max speed for animation 
paddle_A.shape("square") #built in shape square
paddle_A.color('white')
paddle_A.shapesize(stretch_wid=5,stretch_len=1) 
paddle_A.penup() #turtle wont draw while moving 
paddle_A.goto(-350,0) 

# Ball 
ball=turtle.Turtle() 
ball.speed(0) 
ball.shape("circle") 
ball.color('black') 
ball.penup()
ball.goto(0,0)
ball.dx=0.5 #everytime ball moves it moves 2 pixels 
ball.dy=0.5

#pen 
pen=turtle.Turtle()
pen.speed(0)
pen.color('white')
pen.penup() #do not want to draw lines when pen moves 
pen.hideturtle()
pen.goto(0,260)
pen.write("Player Score: 0", align='center', font=('Courier',24,'normal'))

# functions for paddle movement
def paddle_A_up():
    y=paddle_A.ycor() #returns y coordinate assigns variable 
    y+=20 #moves y up 20 pixels 
    paddle_A.sety(y) #updates paddle y 
    
def paddle_A_down():
    y=paddle_A.ycor() 
    y-=20 
    paddle_A.sety(y)  
    
#keyboard binding 
window.listen() #listne for keyboard input
window.onkeypress(paddle_A_up,"w") #when user presses w call function (does not work in caps lock)
window.onkeypress(paddle_A_down,"s") #moves down when user presses 

#main loop for game 
while True: 
    window.update()
    ball.setx(ball.xcor()+ball.dx)
    ball.sety(ball.ycor()+ball.dy)
    
    if ball.ycor()>290: #prevents ball from going offscreen
        ball.sety(290)
        ball.dy *= -1 #reverses 
        
    if ball.ycor()< -290: #prevents ball from going offscreen
        ball.sety(-290)
        ball.dy *= -1 #reverses direction
    
    if ball.xcor()> 390: 
        ball.dx*= 1
        score_A+=1
        pen.clear()
        pen.write("Player Score: {}".format(score_A), align='center', font=('Courier',24,'normal'))
        
    if ball.xcor()< -390:
        ball.dx*= -1
        score_A=0
        pen.clear()
        pen.write("Player Score: {}".format(score_A), align='center', font=('Courier',24,'normal'))
        
    #paddle and ball collisions  
    #paddle A
    if ball.xcor() > 390:
        ball.setx(350)
        ball.dx *= -1 #moves ball opposite direction
        
    elif ball.xcor() < -340 and ball.ycor() < paddle_A.ycor() + 50 and ball.ycor() > paddle_A.ycor() - 50:
        ball.setx(-340)
        ball.dx *= -1 #moves ball opposite direction 
```
